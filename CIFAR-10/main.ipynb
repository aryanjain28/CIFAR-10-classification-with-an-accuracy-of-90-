```python
from tkinter import *
from PIL import ImageTk, Image 
from tkinter import filedialog 
from tkinter import messagebox
import keras
from keras.preprocessing.image import image
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import load_img
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use('dark_background')

fileName = 'xxx'
model = keras.models.load_model('./myModel-elu-activation.h5')

root = Tk() 
root.title("Image Loader") 
root.resizable(width = True, height = True) 
root.minsize(1000, 800)

def open_img():
    x = openfilename() 
    img = Image.open(x)
    
    print(img.size)
    
    W = img.size[0]
    H = img.size[1]
    
    if W > 600:
        H = (600*H)//W
        W = 600
        
    if W < 100:
        H = (100*H)//W
        W = 100
        
    print(W,H)
    
                    
    img = img.resize((W,H), Image.ANTIALIAS) 
    img = ImageTk.PhotoImage(img) 
    panel = Label(root, image = img) 
    panel.image = img 
    panel.place(relx=0.5, rely=0.5, anchor=CENTER)

    
def openfilename():
    global fileName
    fileName = filedialog.askopenfilename(title ='Browse images') 
    return fileName

def imageInfo():
    global fileName
    print(fileName)
    if fileName == 'xxx':
        print('Please upload image')
        messagebox.askokcancel("Error : 404", 'Please upload image')
    else:
        image = load_img(fileName, target_size=(32, 32))
        image = img_to_array(image)
        image = image.reshape((1,32,32,3))
        
        names = ['Airplane','Automobile','Bird','Cat','Deer','Dog','Frog','Horse','Ship','Truck']

        image = image / 255.
        
        result = model.predict(image)
        print(result)
        print(*result)
        print(names[np.argmax(*result)])
                
#         messagebox.askokcancel("Result", names[np.argmax(*result)])
        
        plt.figure(figsize=(10,8))
        plt.bar(names, result[0])
        plt.title(f'\n\nPrediction : {names[np.argmax(*result)]} | Accuracy : {result[0][np.argmax(*result)]*100}% \n ', fontsize=15)
        plt.xticks(rotation=45)
        plt.show()
        
btn1 = Button(root, text ='Upload Image', font=('Ariel', 15), command = open_img) 
btn2 = Button(root, text ='Predict', font=('Ariel', 15), command=imageInfo)

btn1.place(relx=0.5, rely=0.02, anchor=N)
btn2.place(relx=0.5, rely=0.95, anchor=S)

root.mainloop() 
```
